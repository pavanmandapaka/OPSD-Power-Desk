```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
logs = pd.read_csv('../outputs/DE_online_updates.csv', parse_dates=['timestamp'])
forecasts = pd.read_csv('../outputs/DE_live_forecasts.csv', parse_dates=['timestamp']).set_index('timestamp')

drifts = logs[logs['event']== 'DRIFT_ADAPTATION']

if not drifts.empty:
    drift_time = drifts.iloc[0]['timestamp']

    before = forecasts.loc[drift_time - pd.Timedelta(hours=24) : drift_time]
    after = forecasts.loc[drift_time : drift_time + pd.Timedelta(hours=24)]

    mae_before = (before['y_true'] - before['yhat']).abs().mean()
    mae_after = (after['y_true'] - after['yhat']).abs().mean()

    print(f"Drift Event at: {drift_time}")
    print(f"MAE 24h Before: {mae_before:.2f}")
    print(f"MAE 24h After:  {mae_after:.2f}")
    print(f"Improvement:    {mae_before - mae_after:.2f}")
```
