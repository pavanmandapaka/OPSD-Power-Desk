```python
import pandas as pd 
import numpy as np 
import os
```

```python
COUNTRY = 'DE'
PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), '..'))
INPUT_PATH = os.path.join(PROJECT_ROOT, 'outputs', f'{COUNTRY}_anomalies.csv')
```

```python
df = pd.read_csv(INPUT_PATH, parse_dates=['timestamp'], index_col='timestamp')
```

```python
potential_pos = df[df['z_score'].abs() >= 3.5]
potential_neg = df[df['z_score'].abs() < 1.0]
```

```python
n_samples = 50

sample_pos = potential_pos.sample(n=min(len(potential_pos), n_samples), random_state=42)
sample_neg = potential_neg.sample(n=min(len(potential_neg), n_samples), random_state=42)
```

```python
sample_pos = sample_pos.copy()
sample_pos['label'] = 1

sample_neg = sample_neg.copy()
sample_neg['label'] = 0
```

```python
df_to_verify = pd.concat([sample_pos, sample_neg]).sample(frac=1, random_state=42)
```

```python
output_path = os.path.join(PROJECT_ROOT, 'outputs', 'anomaly_labels_to_verify.csv')
df_to_verify.to_csv(output_path)

print(f"Created {output_path}.")
```
