```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df_de = pd.read_csv('../data/DE.csv', parse_dates=['timestamp'], index_col='timestamp')

df_de.iloc[-14*24:]['load'].plot(figsize=(15, 5), title='DE Load - Last 14 Days')
plt.show()
```
